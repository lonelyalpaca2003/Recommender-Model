```python
import pandas as pd 
import numpy as np 
from sklearn.metrics.pairwise import cosine_similarity
import pickle
```

```python
df = pd.read_csv('../data/indian_songs.csv')
df
```

```python
df = df.drop('Unnamed: 0', axis =1) 
```

```python
df = df.sample(15000).reset_index(drop=True)
```

```python
x = ['popularity', 'year','danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo',
       'duration_ms', 'time_signature']

similarity_matrix = cosine_similarity(df[x])
```

```python
def recommender_system(song_name, artist_name=None, n_recommendations=10):
    if song_name in df['track_name'].values:
        song_idx = df[df['track_name'] == song_name].index[0]
        sim_scores = similarity_matrix[song_idx]
        sorted_indices = sim_scores.argsort()[::-1]
        recommended_songs = sorted_indices[1:n_recommendations+1] 
        song_names = df.iloc[recommended_songs][['track_name']]
        print(f'Similar songs you might like include: \n {song_names.to_string(index=False, header=False)}')
    else:
        print("No such song found in the Database, please enter a valid song")
```

```python
np.save('../data/similarity_matrix.npy', similarity_matrix)
```
